```python
# followed the direction of this video https://www.youtube.com/watch?v=zRMmiBMjP9o
def nf_kb(t, x):
    # x is here to be used by (1) initial conditions and (2) x when we use odeint
    # t is here for obvious reasons
    
    # define the k values
    k1=0.2
    k2=0.1
    k3=1
    k4=1
    k5=0.1
    k6=1
    k7=1
    k8=0.182
    k9=0.0182
    k10=0.05
    k11=0.0909
    k12=0.267
    k13=1
    k14=120
    k15=206
    k16=0.5
    k17=0.417
    k18=8.22*10**-5
    k19=0.167
    k20=1
    k21=0.423
    k22=0.000257
    k23=2
    k24=0.02
    k25=0.1
    k26=0.1
    k27=0.1
    
    # im lazy and make it easy to read
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    x5 = x[4]
    x6 = x[5]
    x7 = x[6]
    x8 = x[7]
    x9 = x[8]
    x10 = x[9]
    x11 = x[10]
    x12 = x[11]
    x13 = x[12]
    x14 = x[13]
    x15 = x[14]
    x16 = x[15]
    x17 = x[16]
    x18 = x[17]
    x19 = x[18]
    x20 = x[19]
    x21 = x[20]
    x22 = x[21]
    x23 = x[22]
    x24 = x[23]
    x25 = x[24]
    x26 = x[25]
    
    dx1dt = k1-k2*x1-k3*x1*x2
    dx2dt = -k3*x1*x2-k4*x2*x4
    dx3dt = k3*x1*x2
    dx4dt = -k4*x2*x4-k6*x4*x8
    dx5dt = k4*x2*x4
    dx6dt = -k5*x6*x7-k7*x6*x10
    dx7dt = -k5*x6*x7
    dx8dt = k5*x6*x7-k6*x4*x8
    dx9dt = k6*x4*x8-k8*x9*x12+k8*x9*x12
    dx10dt = -k7*x6*x10-k13*x20*x10+k18-k19*x10+k24*x23
    dx11dt = k7*x6*x10
    dx12dt = -k8*x9*x12+k9*x13
    dx13dt = k8*x9*x12-k9*x13-k10*x25*x13+k10*x25*x13
    dx14dt = k10*x25*x13-k11*x14*x15+k13*x20*x10
    dx15dt = k10*x25*x13-k11*x14*x15
    dx16dt = k12*x25-k14*x17*x16+k16*x18
    dx17dt = -k14*x17*x16-k20*x20*x17+k21-k22*x17-k23*x22*x17
    dx18dt = k15*x26-k16*x18
    dx19dt = k16*x18-k17*x19
    dx20dt = -k13*x20*x10-k20*x20*x17
    dx21dt = k20*x20*x17
    dx22dt = -k23*x22*x17
    dx23dt = k23*x22*x17-k24*x23-k27*x23
    dx24dt = k25-k26*x24+k27*x23
    dx25dt = -k10*x25*x13+k11*x14*x15-k12*x25
    dx26dt = k14*x17*x16-k15*x26
    
    return [dx1dt, dx2dt, dx3dt, dx4dt, dx5dt, dx6dt, dx7dt, dx8dt,
            dx9dt, dx10dt, dx11dt, dx12dt, dx13dt, dx14dt, dx15dt, dx16dt, 
            dx17dt, dx18dt, dx19dt, dx20dt, dx21dt, dx22dt, dx23dt, dx24dt, dx25dt, dx26dt]
# first run for the initial conditions
T = initial_conditions_N
x0x0 = [0 for i in range(26)]
x0x0[3] = 0.1
t = [i for i in range(T+1)]
init_conditions = solve_ivp(nf_kb, (min(t), max(t)), x0x0, t_eval=t).y[:, -1] #only want the last values
```

```python
# here is the state space model

ssm = []
ssm.append(sp.sympify('k1-k2*x1-k3*x1*x2'))
ssm.append(sp.sympify('-k3*x1*x2-k4*x2*x4'))
ssm.append(sp.sympify('k3*x1*x2'))
ssm.append(sp.sympify('-k4*x2*x4-k6*x4*x8'))
ssm.append(sp.sympify('k4*x2*x4'))
ssm.append(sp.sympify('-k5*x6*x7-k7*x6*x10'))
ssm.append(sp.sympify('-k5*x6*x7'))
ssm.append(sp.sympify('k5*x6*x7-k6*x4*x8'))
ssm.append(sp.sympify('k6*x4*x8-k8*x9*x12+k8*x9*x12'))
ssm.append(sp.sympify('-k7*x6*x10-k13*x20*x10+k18-k19*x10+k24*x23'))
ssm.append(sp.sympify('k7*x6*x10'))
ssm.append(sp.sympify('-k8*x9*x12+k9*x13'))
ssm.append(sp.sympify('k8*x9*x12-k9*x13-k10*x25*x13+k10*x25*x13'))
ssm.append(sp.sympify('k10*x25*x13-k11*x14*x15+k13*x20*x10'))
ssm.append(sp.sympify('k10*x25*x13-k11*x14*x15'))
ssm.append(sp.sympify('k12*x25-k14*x17*x16+k16*x18'))
ssm.append(sp.sympify('-k14*x17*x16-k20*x20*x17+k21-k22*x17-k23*x22*x17'))
ssm.append(sp.sympify('k15*x26-k16*x18'))
ssm.append(sp.sympify('k16*x18-k17*x19'))
ssm.append(sp.sympify('-k13*x20*x10-k20*x20*x17'))
ssm.append(sp.sympify('k20*x20*x17'))
ssm.append(sp.sympify('-k23*x22*x17'))
ssm.append(sp.sympify('k23*x22*x17-k24*x23-k27*x23'))
ssm.append(sp.sympify('k25-k26*x24+k27*x23'))
ssm.append(sp.sympify('-k10*x25*x13+k11*x14*x15-k12*x25'))
ssm.append(sp.sympify('k14*x17*x16-k15*x26'))
```

```python
# get Sj. ONLY FOR INITIAL CONDITIONS OF SOLVING THE NEXT SYSTEM OF EQUATIONS
def kdf(i, j):
    if i != j:
        return 0
    return ks['k{}'.format(j+1)] - x0[i]

def sj(j):
    return [kdf(i, j) for i in range(len(x0))]

# get Fj (df/dthetaj)
def fj(j):
    return [sp.diff(ssm[i], 'k{}'.format(j+1)) for i in range(len(x0))]

# find J
def makeJ():
    J = []
    # df1 is top row where as dx1 is the first column
    # is dfn/dfn
    for i in range(len(ssm)):
        row = []
        tod = ssm[i]
        for n in range(len(ssm)):
            respectto = 'x{}'.format(n+1)
            row.append(sp.diff(tod, respectto))
        J.append(row)
    return J

J = makeJ()
def ddtdxdthetaj(j, p):
    return np.add(np.dot(np.asarray(J), p), np.asarray(fj(j)))

# linear interpolation for x 
# x is i by N shape, so index i then t
thresh = .000001
def interpx(i, t):
    if abs(t - int(t)) < thresh:
        return x[i][int(t)]
    t1 = int(math.floor(t))
    t2 = int(math.ceil(t))
    x1 = x[i][t1]
    x2 = x[i][t2]
    return ((t-t1) * (x2 - x1)/(t2-t1)) + x1

```

```python
import pickle
from multiprocessing import Process, Manager

def integratethis(t, p, soe):
    tosub = {}
    tosolve = copy.deepcopy(soe)
    for i in range(len(x0)):
        tosub['x{}'.format(i + 1)] = interpx(i, t)
    for k in ks:
        tosub[k] = ks[k]
    for pi in range(len(p)):
        tosub['p{}'.format(pi+1)] = p[pi]
    for j in range(len(tosolve)):
        tosolve[j] = float(tosolve[j].evalf(subs=tosub))
    return list(tosolve)
filename = 'integrated_{}'
N = 400
t = [i for i in range(N)]
# the form of this solved system will be (j, N, i) so first index is the theta, the second is the time, and the third is the x
solveddxdtheta = []
    
def solveit(j, soe, l):
    p0 = sj(j)
    solved = solve_ivp(integratethis, (min(t), max(t)), p0, args=(soe, ), t_eval=t).y
    l.append((j, solved))
    with open(filename.format(j), 'wb') as o:
        pickle.dump(solved, o)
    print('finished j={}'.format(j))
    
with Manager() as manager:
        L = manager.list()  # <-- can be shared between processes.
        processes = []
        js_to_do = [15, 17, 19, 21, 24, 25]
        for j in js_to_do:
            soe = ddtdxdthetaj(j, [sp.sympify('p{}'.format(i + 1)) for i in range(len(x0))])
            p = Process(target=solveit, args=(j, soe, L))  # Passing the list
            p.start()
            processes.append(p)
        for p in processes:
            p.join()
        print(L)
```
